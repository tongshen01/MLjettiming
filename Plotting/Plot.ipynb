```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
Double_A_norboth = np.load('Double_A_norboth.npy')
Double_A_unnorboth = np.load('Double_A_unnorboth.npy')
Double_A_cat = np.load('Double_A_cat.npy')
A_pT = np.load('A_pT_double.npy')
```

```python
plt.hist(Double_A_cat,bins=np.linspace(-0.05,0.05,80),density=True,histtype="step",label="cat")
plt.hist(Double_A_norboth,bins=np.linspace(-0.05,0.05,80),density=True,histtype="step",label="nor")
plt.hist(Double_A_unnorboth,bins=np.linspace(-0.05,0.05,80),density=True,histtype="step",label="unnor")
plt.hist(A_pT,bins=np.linspace(-0.05,0.05,80),density=True,histtype="step",label="$p_T$ weight",ls=":")
plt.yscale("log")
plt.xlabel("relative time difference")
plt.ylabel("fraction per bin")
plt.legend(frameon=False)
```
